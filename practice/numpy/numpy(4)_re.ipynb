```python
import re

# 파일을 열고 내용을 읽은 다음, 'better'를 'butter'로 대체합니다.
with open('./test.txt', 'r+', encoding='UTF-8') as f:
    contents = f.read(-1)
    tokens = contents.split(' ')
    for i in range(len(tokens)):
        if re.match(r'[.+?]?better[.+?]?', tokens[i]):
            tokens[i] = "butter"
    print(tokens)
    updated_contents = ' '.join(tokens)
    
    # 파일 포인터를 시작 위치로 이동합니다.
    f.seek(0)
    # 수정된 내용을 파일에 씁니다.
    f.write(updated_contents)
    # 파일의 나머지 부분을 삭제합니다(수정된 내용이 원래 내용보다 짧을 경우를 대비).
    f.truncate()
```

```python
def readfile(filepath='auto.txt'):
    if filepath:
        with open(filepath,'r',encoding='utf-8') as f:
            return f.readlines()
```

```python
def resetfile(filepath='auto.txt'):
    if filepath:
        with open(filepath,'w',encoding='utf-8') as f:
            return f.truncate()
```

```python
def rewritefile(filepath='auto.txt', contents=""):
    if filepath:
        with open(filepath,'w',encoding='utf-8') as f:
            return f.write(contents)
```

```python
def writefile(filepath='auto.txt', contents=""):
    if filepath:
        with open(filepath,'a',encoding='utf-8') as f:
            return f.write(contents)
```
