```python
# 딕셔너리에서 키는 중복안된다. 키가 중복되면 재할다이 발생한다.
# 딕셔너리에서는 키가 중요하다,
# 딕셔너리의 __iter__ 는 기본적으로 키를 뽑아낸다.
for i in {'a':1, 'b':2, 'c':3 }:
  print(i)


# 딕셔너리 뷰
for i in {'a':1, 'b':2, 'c':3 }.keys():
  print(i)

for i in {'a':1, 'b':2, 'c':3 }.values():
  print(i)

for i in {'a':1, 'b':2, 'c':3 }.items():
  print(i)


# 고급 사용자 (for의 비밀)
class A:
  a = 1
  def __iter__(self):
    return iter([1,2,3])

aa = A()

for i in aa:
  print(i)


# while은 내일
# 수요일 부터 어려운것
```
