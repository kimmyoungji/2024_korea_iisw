```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.groupby('sex')['tip'].mean().plot.bar()
plt.title('title')
plt.xticks('M','F')
```

```python
plt.grid('False') # state machine 방식이기 때문에, -> matplotlib의 1버전
```
