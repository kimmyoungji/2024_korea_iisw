```python
import matplotlib.pyplot as plt
```

# matplotlib 버전1: state machine 방식

```python
# state machine
1. 변수를 지정하지 않음. 가장 가가이 있는 것을 찾음
2. 계층적 디폴트 값 자동 생성
```

## matplotlib는 그래프를 그릴 가상 공간을 생성한다. == Figure

```python
# plt.figure();
plt.figure(figsize=(10,5))
```

Signature:
plt.figure(
    num: 'int | str | Figure | SubFigure | None' = None,
    figsize: 'tuple[float, float] | None' = None,  # 가상의 공간의 크기 지정 # Axes에서 실제로 그린다.
    dpi: 'float | None' = None,
    *,
    facecolor: 'ColorType | None' = None,
    edgecolor: 'ColorType | None' = None,
    frameon: 'bool' = True,
    FigureClass: 'type[Figure]' = <class 'matplotlib.figure.Figure'>,
    clear: 'bool' = False,
    **kwargs,
) -> 'Figure'
Docstring:
Create a new figure, or activate an existing figure.

Parameters

```python
# plt.figure 없으면 자동 생성. 기본으로.
plt.axes([0,0.5,1,1])  
plt.plot([1,2,3,4]) # 위로 찾아서 가장 가까운 axes에 그려진다.  # 원하는것을 아주 빨리 그릴 수 있다.  # 아주 공돌스럽다. # 나는 ggplot 방식이 더 궁금하다.
# plt.grid(True)
plt.axes([0.5,0.5,0.5,1])
plt.plot([2,4,5,4])
plt.grid(True)
```

Signature: plt.axes(arg: 'None | tuple[float, float, float, float]' = None, **kwargs) -> 'matplotlib.axes.Axes'
Docstring:
Add an Axes to the current figure and make it the current Axes.

Call signatures::

    plt.axes()
    plt.axes(rect, projection=None, polar=False, **kwargs)   # 영역을 잡을 수 있다,
    plt.axes(ax)

Parameters

```python
plt.axes([0,0.5,1,1])
plt.plot([10,20,30])
```

```python
plt.axes([1,1.5,2,1])
plt.plot([1,2,3,4,5])
```

```python
# plt.figure() 자동으로 생성된다.   # 뭔가 만들 때, 바닥부터 자동으로 만들어주는 것 == state machine의 특성
plt.axes([0.5,0.5,1,1])  # 반시계 방향
```

# style 적용하기

```python
styles = plt.style.available
```

```python
plt.style.use('classic')
```

```python
plt.axes([0.5,0.5,0.5,1])
plt.plot([2,4,5,4])
plt.grid(True)
```

```python
plt.plot([1,2,3,4],[5,6,7,8])
```

```python
plt.axes(polar=True)
plt.plot ([1,2,3,4])
```

```python
plt.barh([1,2,3,4],[5,6,7,8])
plt.title('sun')
plt.xlabel('sun')
plt.ylabel('sun')  # => 자기 만의 스타일을 만들 수 있다. 
plt.xticks([2,3])
plt.yticks([4,1.5])
plt.barh([1,2,3,4],[5,6,7,8])
```
