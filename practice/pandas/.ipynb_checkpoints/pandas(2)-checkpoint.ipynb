```python
mpg.duplicated().value_counts()
```

```python
iris = sns.load_dataset('iris')
```

```python
iris.duplicated(keep='last').value_counts()
```

```python
# 도메인 지식이 없기 때문에, subgrouping으로 문자열의 오류를 잡아낸다.
# 숫자는 범위로 subgrouping을 한다. -> 불린인덱싱
```

```python
iris[iris.duplicated(keep='last')]
```

```python
# 중복제거
iris.drop_duplicates()
```

```python
# 칼럼의 일부만 중복되는 경우
```

```python
iris[iris.iloc[:,:-1].duplicated(keep=False)]
```
