# Map의 단점
- 맵은 1차 데이터만 적용할 수 있다.
- 시리즈만 맵을 적용할 수 있다.
---
- pandas는 딕셔너리까지 맵을 지원

# Transform == 2차원 배열를 위한 맵
- 한번에 적용되는 것 처럼 보이지만, 요소요소 적용한다.

```python
tips.tip
```

```python
tips[['tip','size']].groupby('size').transform(lambda x: x+1)
```

# applymap은 dataframe에 적용
- Transform은 group 형태의 객체, applymap은 dataframe에 적용


```python
tips.select_dtypes('float64').applymap(lambda x:x+1)  # 데이터 영역별로 한번에 수정할 수 있다.
                                                      # applymap 함수만 잘 만들면, 상황에 따라 그에 맞게 데이터 수정가능
```
