```python
# 화면에 안보이면, 보일 때 까지 기다려야한다.
# 매크로 만들 때 이런 요소가 보이지 않는 문제가 발생한다. 
```

```python
# wait != sleep()
explicit_wait => 누군가(명확히 대상)가 어떤 이벤트(오류)가 사라질 때까지 기다리는 것
implicit_wait => 암묵적으로 기다리는 것
```

```python
driver.implicitly_wait = 10  # sleep 을 대신 사용하면 안된다. 뒷단에서 계속 돌아가고 있는 것이다.
```

```python
from selenium.webdriver.support import expected_conditions as ec
from selenium.webdriver.support.wait import WebDriverWait
```

```python
wait = WebDriverWait(driver, 10, 0.1, [ec.visibility_of, ec.element_to_be_clickable])  # 드라이버, 몇초동안, 반복주기
```

```python
from selenium.webdriver.common.by import By
a = driver.find_elements(By.CSS_SELECTOR, 'tr .dropdown-item')
```

```python
# Selenium => Browser => Render Tree => ... => 
# sleep X, wait(Explicit => 대상의 상태를 기다려주는 것, Implicit => 로드)
driver.implicitly_wait
```

```python
wait.until(ec.visibility_of(a))
wait.until(ec.element_to_be_clickable(a))
```
