```python
# 띄어쓰기
def ngram(s, n=2, t=True): # t=True;어절, False;음절
    result = []
    
    if not t:
        s = list(s)
        
    for i in range(len(s)-(n-1)):
        result.append(tuple(s[i:i+n])) # 튜플로 수정
        
    return result
```

```python
corpus = list()
for file in fileids('news'):
    if file == "news/.ipynb_checkpoints":
        continue
    with open(file, 'r', encoding='utf8') as f:
        corpus.append(
            re.sub(r'^\s+|\s+$', '',
                   re.sub(r'\s+', ' ',
                          re.sub(r'\sCopyright.+', ' ',
                                 re.sub(r'[\xa0-\xff]', ' ', f.read())))))
```

```python
gram = {2:[],3:[],4:[],5:[] }
for c in corpus:
    for s in sent_tokenize(c):
        gram[2].extend(ngram(s,2,False))
        gram[3].extend(ngram(s,3,False))
        gram[4].extend(ngram(s,4,False))
        gram[5].extend(ngram(s,5,False))

# 길이 n인 어절의 집합. - "abcde.." bigram이면 ab,bc,cd,de...
print(gram[2])
```

# 어간 분리

```python
# Branch Entropy => stemming
# -sum(p*log p)
# p=> P(x_i|x_0 ... x_i-1)
# -sum(p*log p)
tokens = Counter('\n'.join(corpus).split()) # 쪼개고, 갯수 새서, { 빈도:어절...}형태로
```

```python
from math import log
from collections import defaultdict

# 어근 분리 -> 첫번째 글자 부터 다음 에 올수 있는 글자 사이의 엔트로피(변화가능성)을 바탕으로 어간 분리 
def stemming(q):
    threshold = 10.0
    # q를 이용하여 각 단계마다 얼마마늠의 엠비규어티를 본다.
     # print('찾으려는 대상 문장:  ', q)
    for i in range(len(q)-1):
        candidates = defaultdict(lambda:0)
        
        # print('keys:  ', f'{keys[:10]}...')
        # print('찾으려는 키1:  ', q[:i+1])
        keys = [ k for k in tokens.keys() if re.match(q[:i+1], k)]  # k로 시작하는 모든 글자들의 빈도수 tokne에 이게 얼마나있는지 본다.
        given = sum([tokens[k] for k in keys])
        # print('given:  ',given, ": 찾은 키들의 빈도수 총합")

        # print('찾으려는 키2:  ', q[:i+2])
        # print('keys2:  ', f'{keys2[:10]}...')
        # keys2 = [ k for k in tokens.keys() if re.match(q[:i+1], k)]  # k로 시작하는 모든 글자들의 빈도수 tokne에 이게 얼마나있는지 본다.
        
        for k in [ k for k in tokens.keys() if re.match(q[:i+1], k)]:
            candidates[k[:i+2]] += tokens[k]
            
        be = -sum([(v/given)*log(v/given) for k,v in candidates.items()])
        print(be)

        # 엔트로피가 올라가는 경우를 캐치
        if be > threshold:
            # print(i+1, q[:i+1], q[i+1:])
            break

        threshold = be
    
        return (q,) if i == len(q)-1 else (q[:i+1], q[i+1:])
        
stemming('금융감독원')
```
