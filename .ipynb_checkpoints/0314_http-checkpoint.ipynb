```python
# robots.txt 확인
# 400번대 500번대 분기
```

```python
from requests.compat import urlparse

def canFetch(url, path):
    # disallowList = list()
    resp = get(urljoin(url, '/robots.txt'))
    
    if resp.status_code != 200:
        return True

    disallowList = [link for link in re.findall(r'Disallow\s*:\s*(.+)',resp.text, re.IGNORECASE | re.MULTILINE)]

    if urlparse(path).path in disallowList:
        return False

    return True

canFetch('https://lms.sunde41.net','/')
        
    
```

```python
canFetch('https://www.google.com','/search?q=수지')
```

```python
from requests import request
from requests.exceptions import HTTPError
from time import sleep

def download(url, params={}, data={}, headers={}, method="GET", retries=3):
    if not canFetch(url, url):
        print('수집하면 안됨')
    resp = request(method, url, params=params, data=data, headers=headers)
    try:
        resp.raise_for_status()
    except HTTPError as e:
        if 499 < resp.status_code and retries > 0:
            print('재시도 중')
            sleep(5)
            return download(url, params, data, headers, method, retries=1 )
        else: 
            print(e.response.status_code)
            print(e.request.headers)
            print(e.response.headers)
            return None
    return resp
```

```python
## 크롤링
웨크롤링: 시스템적으로 브라우저를 통해 웹페이지를 보는것, 목적은 인덱싱
웹인덱싱 : 어떤 웹페이지가 무슨 단어를 사용하고, 어떤 맥락에서 단어를 사용하는지에 따라 페이지 분류

## 구조
PPT 이미지

# 구글도 웹도 30% 밖에 크롤링 못함. 70%가 다크웹,다크웹 그냥 크롤러가 못간 곳임.

# Scrapy vs BeautifulSoup 구별하기 (?)
```

```python
# 웹 크롤러 만들기
1. ㅣink를 찾아야됨 => link
    a[href], from[action], iframe[src],/ img[src], video/audro[src], css/js[src] 
                      여기까지는 크롤링해야함|
2. 찾은 링크를 따라다니게 만들어야함. (?) => 전략
    WWW => X (Unknown search space) => 깊이우선탐색(DFS), 너비우선탐색(BFS) 트리탐색
    DFS - 찍고오는것 - stack: FILO   # 사이트의 끝을 보고 싶을 때 좋다. 그러나, 불균형트리모양의 사이트일 경우, 언제 끝날지 알수 없다.
    BFS - 순차적으로 - Queue: FIFO   # 관련있는 것들이 나올 확률이 높다. 그러나, 메모리르 계속 들고 있어야함
            root
        a          b 
     c     d    e     f 
   DFS: rootabcdef
   BFS: rootabefcd
3. 웹이 어디서 시작하고, 어디서 끝나는지 알수 없어서 DFS와 BFS가 필요하다.
    
4. 새롭게 찾은 link 검사(absolute path) 정규화 => 규칙
javascript:, #, ./, ../ => urljoin으로 원해 주소랑 결합하자.
5. 새롭게 찾은 link 방문한적이 있는지 검사 => 관리
seen? => list => DB(ORM)


iframe(다른 사이트에 있는 것을 마치 나의 것처럼), 
```

```python
URLs = ['http://inisw.kr']
```

```python
Visited = []
```

```python
while URLs:
    url = URLs.pop(-1) # 0:FIFO:Queue, -1:LIFO:STACK

    Visited.append(url)

    # robots.txt를 확인하여 크롤링 허가여부 확인 후 웹 페이지 텍스트 가져오기
    resp = download(url)

    # 만약 응답 값이 비어있다면 예외처리하기
    if resp is None:
        continue

    # 만약 응답값이 텍스트 타입 데이터라면 dom으로 변환하기
    if re.search(r'text\/html', resp.headers['content-type']):
        dom = BeautifulSoup(resp.text, "html.parser")

        # 링크가 있을 것으로 예상되는 요소 선택하기
        for link in dom.select('''a[href], frame[src], iframe[src], img[src],
        audio[src], video[src], style[src], link[src]'''): # form['action']은?
            
            # 찾은 링크의 url을 추출하여 newURL에 담기
            newURL = link.attrs['href'] if link.has_attr('href') else link.attrs['src']

            # 만약 url이 # 또는 javascript 또는 mailto가 아니라면, 진행
            if not re.match(r'#|javascript|mailto', newURL):
                # url 정규화
                normalizedURL = urljoin(url, newURL)
                # 방문한 url인지 확인하기
                if normalizedURL not in Visited and \
                   normalizedURL not in URLs:         # 이미 등록된 url인지 확인하기
                       URLs.append(normalizedURL)
                    
    print(len(URLs), len(Visited))
```
